```python
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem.AllChem import GetMorganFingerprint
from rdkit.Chem.Draw import DrawMorganBit
```

```python
mol = Chem.MolFromMolFile('db_mol/495.mol')
mol
```

```python
radius = 3

bi = {}
ecfp = GetMorganFingerprint(mol, radius=radius, bitInfo=bi, useFeatures=False, useBondTypes=False)
# 由于openbabel无法准确判断键的类型，须设置 useBondTypes=False

col = 5
row = (len(bi) - 1) // col + 1
fig, axes = plt.subplots(row, col, figsize=(col * 3, row * 3), tight_layout=True)
for ax in axes.ravel():
    ax.axis('off')
for ax, bit in zip(axes.ravel(), bi.keys()):
    ax.imshow(DrawMorganBit(mol, bit, bi))
#     ax.set_title(bit)
```
